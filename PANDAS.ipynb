MERGING, JOINING AND CONCATENATING THE DATA FRAMES

```python
df1 = pd.read_csv(r"C:\Users\sunny\Desktop\DATA ANALYSIS\PYTHON\Pandas\LOTR.csv")
df1
```

```python
df2 = pd.read_csv(r"C:\Users\sunny\Desktop\DATA ANALYSIS\PYTHON\Pandas\LOTR 2.csv")
df2
```

```python
df1.merge(df2)        #joining two df's

#here by default df1 will be considered as right circle while df2 as left 
#and also by default it will return inner join as we didn't mention which join we want to return
```

```python
df1.merge(df2, how = 'inner', on = 'FellowshipID')

#it returns the innerjoin based on the 'FellowshipID' so it gives you FirstName_x and FirstName_y
```

```python
df1.merge(df2, how = 'inner', on = ['FellowshipID', 'FirstName'])

#here we merge on fellowshipid along with firstname so that only one combined firstname column will be returned
```

```python
df1.merge(df2, how = 'outer')

#outerjoin
```

```python
df1.merge(df2, how = 'left')

#it returns the everything in left df which matches to the right df
```

```python
df1.merge(df2, how = 'right')

#it returns the everything in right df which matches to the left df
```

```python
df1.merge(df2, how = 'cross')

#it rarely used cases
```

```python
df1.join(df2, on = 'FellowshipID', how = 'outer', lsuffix = '_Left',rsuffix = '_Right')

#in joins we must need to specify the suffix properly otherwise it returns error and also we can't just join df's without mention on which column
```

```python
df4 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'), lsuffix = '_Left',rsuffix = '_Right', how = 'outer')
df4

#here we are setting the index to both the df's then we r joining
```

```python
pd.concat([df1,df2])

#concatinating the df's
```

```python
pd.concat([df1,df2], join = 'inner')

#these 2 columns are similar so it returns
```

```python
pd.concat([df1,df2], join = 'outer')
```

```python
pd.concat([df1,df2], join = 'outer', axis = 1)
```

```python
df1.append(df2)
```

DATA CLEANING IN PANDAS

```python
df=pd.read_excel(r"C:\Users\sunny\Desktop\DATA ANALYSIS\PYTHON\Pandas\Customer Call List.xlsx")
df
```

```python
df = df.drop_duplicates()
df

#first we need to remove the duplicates
```

```python
df = df.drop(columns = "Not_Useful_Column")
df

#drop unneccessary columns
```

```python
df["Last_Name"] = df["Last_Name"].str.lstrip("...")
df["Last_Name"] = df["Last_Name"].str.lstrip("/")
df["Last_Name"] = df["Last_Name"].str.rstrip("_")
df

#one way to remove unneccesary things
```

```python
df["Last_Name"] = df["Last_Name"].str.strip("123._/")
df

#another way to remove unneccesary things
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]','')
df

#first we need to replace the column with nothing like it's became just a normal text
```

```python
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x))
df

#now we need to convert the text format into string by applying lamba
```

```python
df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
df

#now we create a format
```

```python
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
df
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace('nan--','')

df["Phone_Number"] = df["Phone_Number"].str.replace('Na--','')
df

#now we remove null's
```

```python
df[["Street_Address", "State", "Zip_Code"]] = df["Address"].str.split(',',2, expand=True)
df

#we need to split the columns
```

```python
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('Yes','Y')
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('No','N')
df

#first we need to change yes to y otherwise it will be like yeses ans noo will be returned
```

```python
df["Paying Customer"] = df["Paying Customer"].str.replace('Yes','Y')
df["Paying Customer"] = df["Paying Customer"].str.replace('No','N')
df

#first we need to change yes to y otherwise it will be like yeses ans noo will be returned
```

```python
#df = df.replace('N/a','')
#df = df.replace('NaN','')

#this will not workout as technically they are empty values

df=df.fillna('')
df
#so we need to fillup them with empty 
```

```python
for x in df.index:
    if df.loc[x, "Do_Not_Contact"] == 'Y':
        df.drop(x, inplace=True)

df

#we need to remove not to call customers numbers so first remove y from 'Do_Not_Contact'
#here loc= index location
```

```python
for x in df.index:
    if df.loc[x, "Phone_Number"] == '':
        df.drop(x, inplace=True)

df

#now removing the spaces from Phone_Number to visible clean
```

```python
#Another way to drop null values
#df = df.dropna(subset="Phone_Number"), inplace=True)
```

```python
df = df.reset_index(drop=True)
df

#now we need to reset the index, we have to write drop=true otherwise old index will be shown as it is
```
