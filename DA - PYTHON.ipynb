10. WHILE LOOP

```python
number = 0

while number < 5:
    print(number)
    number = number + 1
```

```python
number = 0

while number < 5:
    print(number)
    if number == 3:   #it will break the loop after it becomes true
        break
    number = number + 1
```

```python
number = 0

while number < 5:
    print(number)
    if number == 3:
        break
    number = number + 1
else:                            #here else can't work because the loop had alread breaked
    print('No longer < 5')
```

```python
number = 0

while number < 5:
    number = number + 1
    if number == 3:
        continue               #it continues the loop even after it return true in if condition 
    print(number)
else:
    print('No longer < 5')
```

```python
#BMI Calculator

 
#BMI = (weight in pounds x 703) / (height in inches x height in inches)
#condition:

#Under 18.5	Underweight	Minimal
#18.5 - 24.9	Normal Weight	Minimal
#25 - 29.9	Overweight	Increased
#30 - 34.9	Obese	High
#35 - 39.9	Severely Obese	Very High
#40 and over	Morbidly Obese	Extremely High

#code:
name = input("Enter you name: ")

weight = int(input("Enter your weight in pounds: "))

height = int(input("Enter your height in inches: "))

BMI = (weight * 703) / (height * height)

print(BMI)

if BMI>0:
    if(BMI<18.5):
        print(name +", you are underwight.")
    elif (BMI<=24.9):
        print(name +", you are normal weight.")
    elif (BMI<29.9):
        print(name +", you are overweight. You need to exercise more and stop sitting and writing so many python tutorials.")
    elif (BMI<34.9):
        print(name +", you are obese.")
    elif (BMI<39.9):
        print(name +", you are severely obese.")
    else:
        print(name +", you are morbidly obese.")
else:
    print("Enter valid input")

if BMI>0:
    if(BMI<18.5):
        print(name +", you are underwight.")
    elif (BMI<=24.9):
        print(name +", you are normal weight.")
    elif (BMI<29.9):
        print(name +", you are overweight.")
    elif (BMI<34.9):
        print(name +", you are obese.")
    elif (BMI<39.9):
        print(name +", you are severely obese.")
    else:
        print(name +", you are morbidly obese.")
else:
    print("Enter valid input")
```

14. AUTOMATIC FILE SORTER IN FILE EXPLORER:

```python
import os, shutil
path = r"C:/Users/sunny/Desktop/python tutorial/"             #path of one of the element in that folder

#in the above line(path) we need to put r for raw string and also insert backward slash at end of the path and also 
#change all the forward slashes to backward slashes
#the actual path is (C:\Users\sunny\Desktop\python tutorial) we need to change to above path


os.listdir(path)                                       #it returns the all the elements 

os.path.exists(path+'csv files')                     #it checks the folder that we are creating is already exists or not 

folder_names = ['csv files', 'image files', 'text files']        #creating the list of folders

for loop in range (0,3):
 
  if not  os.path.exists(path+folder_names[loop]):          #creating the folders using loop if the folders is not there
      os.makedirs(path+folder_names[loop])

file_name = os.listdir(path) 



for file in file_name:
    if ".csv" in file and not os.path.exists(path+"csv files/"+file):
        shutil.move(path+file,path+"csv files/"+file)
    elif ".jpeg" in file and not os.path.exists(path+"image files/"+file):
        shutil.move(path+file,path+"image files/"+file)
    elif ".jpg" in file and not os.path.exists(path+"image files/"+file):
        shutil.move(path+file,path+"image files/"+file)
    elif ".txt" in file and not os.path.exists(path+"text files/"+file):
        shutil.move(path+file,path+"text files/"+file)
  


```

```python
#BEAUTIFULSOUP AND REQUESTS ARE THE PACKAGES IN PYTHON WHICH ARE USED TO HELP IN WEB SCRAPPING

from bs4 import BeautifulSoup
import requests
```

```python
url = "https://www.scrapethissite.com/pages/forms/"    #giving source from where the data comes from
```

```python
requests.get(url)        


# it returns a response object 
#if it returns "<Response[200]>" thats ok 
#if it returns "<Response[204]>" that means there is no content in actual web page
#if it returns "<Response[400]>" that means a bad request, its invalid server can't process it
#if it returns "<Response[404]>"  means that you're familiar with than's an error that server cannot be found
```

```python
page = requests.get(url)             
soup = BeautifulSoup(page.text, 'html')   #it will return all the code in html
print(soup)
```

```python
print(soup.prettify())         #it will arrange the code in heirarcy(in order specifically with spaces)
```
